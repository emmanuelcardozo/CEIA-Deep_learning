#### Comparar Mini-Batch Gradient Descent vs Orden 1 para entrenar un modelo cuadrático

Para tal fin, se creó un dataset sintético con 10000 muestras a partir de la siguiente expresión (donde $x$ es la feature que toma valores entre 0 y 6, e $y$ es la salida): 

#### $y = -(x - 3)^{2} + 2 + 0.1 sin(10x)$

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n_samples = 10000

f = lambda x: -((x - 3) ** 2) + 2 + 0.1 * np.sin(10 * x)

x = np.random.uniform(low=0, high=6, size=n_samples)
y = f(x)

plt.figure(figsize=(15, 15))
plt.scatter(x, y, marker=".")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

#### Suponer que se tiene un modelo, tal que la relación entre la salida y la entrada está dada por la ecuación presentada a continuación.

#### $\hat{y} = w_{1}x^{2} + w_{2}x + b$

#### Suponiendo que se va a utilizar mini-batch SGD para encontrar los valores óptimos de los pesos $w_{1}$, $w_{2}$ y $b$ tal que minimizan el ECM (Error Cuadrático Medio) (el ECM es la función de costo para este problema), ¿cuál sería la regla de actualización de cada parámetro?. 

Dada la función de costo, el ECM, se define su derivada respecto de $\hat{y}$.

$J(y, \hat{y}) = (y - \hat{y})^{2}$

$\frac{\partial J}{\partial \hat{y}} = -2(y - \hat{y}) $

Se calcula la derivada parcial de $\frac{\partial J}{\partial w_{i}}$ para cada valor de $w$ para obtener la regla de actualización de cada parámetro.

$\frac{\partial J}{\partial w_{1}} = \frac{\partial J}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial w_{1}} = -2(y - \hat{y}) \cdot x^{2}$

$\frac{\partial J}{\partial w_{2}} = \frac{\partial J}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial w_{2}} = -2(y - \hat{y}) \cdot x$

$\frac{\partial J}{\partial b} = \frac{\partial J}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial b} = -2(y - \hat{y})$

Agregando el learning rate $\alpha$ la regla de actualizaion para cada $w$ es la siguiente:

$w_{1}= w_{1} - \alpha \cdot (-2(y - \hat{y}) \cdot x^{2})$

$w_{2}= w_{2} - \alpha \cdot (-2(y - \hat{y}) \cdot x)$

$b = b - \alpha \cdot (-2(y - \hat{y}))$

#### Implementar el algoritmo en Python (usar mini-batch), encontrar los valores óptimos para $w_{1}$, $w_{2}$ y $b$ y reportar la evolución del ECM con el número de epochs.




```python
class Mini_Batch_GD:
    def initializeParameters(self):
        W1, W2, b = np.random.rand(3)
        self.params = (W1, W2, b)

    def fit(self, x, y, epochs, learning_rate, batch=1):
        self.initializeParameters()

        history = []
        n = x.shape[0]

        for epoch in range(epochs):
            for batch_start in range(0, n, batch):
                batch_end = min(batch_start + batch, n)
                self.forward(
                    x[batch_start:batch_end],
                    y[batch_start:batch_end],
                )
                self.backwardPropagation(
                    x[batch_start:batch_end],
                    y[batch_start:batch_end],
                    learning_rate,
                )
            history.append(self.cost.mean())
            if epoch % 100 == 0:
                print(f"Epoch {epoch}: {history[-1]:.4f}")

        return history

    def forward(self, x, y):
        W1, W2, b = self.params
        self.y_hat = W1 * x**2 + W2 * x + b
        self.cost = (y - self.y_hat) ** 2

    def backwardPropagation(self, x, y, learning_rate):
        W1, W2, b = self.params

        dy_hat = -2 * (y - self.y_hat)
        dW1 = dy_hat * x**2
        dW2 = dy_hat * x
        db = dy_hat

        self.params = (
            W1 - learning_rate * dW1.mean(),
            W2 - learning_rate * dW2.mean(),
            b -  learning_rate * db.mean(),
        )

    def predict(self, x):
        W1, W2, b = self.params
        return W1 * x**2 + W2 * x + b
```

Probamos el modelo con puntos equidistantes en el intervalo $[0, 6]$ y un batch size=10

```python
custom_model = Mini_Batch_GD()
history = custom_model.fit(x, y, 1000,0.001, batch=50)

x_test = np.linspace(0, 6, 50)
y_pred = custom_model.predict(x_test)

plt.figure(figsize=(15, 10))
plt.scatter(x, y, label="Data Point", marker=".")
plt.scatter(x_test, y_pred, label="Prediction")
plt.plot(x_test, y_pred, color="k", label="Fitted curve")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

La variación de loss por epoch es la siguiente

```python
plt.figure(figsize=(15, 10))
plt.plot([i for i in range(1, 1001)], history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

#### Se Implementa el algoritmo de First Order Momentum, y se vuelve a  buscar los valores óptimos de $w_{1}$, $w_{2}$ y $b$ para reportar la evolución del ECM con el número de epochs.

```python
class First_Order_Mom:
    def initializeParameters(self):
        W1, W2, b = np.random.rand(3)
        self.params = (W1, W2, b)
        dW1_prev, dw2_prev, db_prev = np.float64(0), np.float64(0), np.float64(0)
        self.previous = (dW1_prev, dw2_prev, db_prev)

    def fit(self, x, y, epochs, learning_rate, sigma_vel, batch=1):
        self.initializeParameters()

        history = []
        n = x.shape[0]

        for epoch in range(epochs):
            for batch_start in range(0, n, batch):
                batch_end = min(batch_start + batch, n)
                self.forward(
                    x[batch_start:batch_end],
                    y[batch_start:batch_end],
                )
                self.backwardPropagation(
                    x[batch_start:batch_end],
                    y[batch_start:batch_end],
                    learning_rate, sigma_vel
                )
            history.append(self.cost.mean())
            if epoch % 100 == 0:
                print(f"Epoch {epoch}: {history[-1]:.4f}")

        return history

    def forward(self, x, y):
        W1, W2, b = self.params
        self.y_hat = W1 * x**2 + W2 * x + b
        self.cost = (y - self.y_hat) ** 2

    def backwardPropagation(self, x, y, learning_rate, sigma_vel):
        W1, W2, b = self.params

        dy_hat = -2 * (y - self.y_hat)
        dW1 = dy_hat * x**2
        dW2 = dy_hat * x
        db = dy_hat
        
        vel_W1, vel_W2, vel_b = self.previous
        
        self.params = (
            W1 - sigma_vel * vel_W1.mean() - learning_rate * dW1.mean(),
            W2 - sigma_vel * vel_W2.mean() - learning_rate * dW2.mean(),
            b -  sigma_vel * vel_b.mean()  - learning_rate * db.mean(),
        )
        
        self.previous = learning_rate*dW1.mean(), learning_rate*dW2.mean(), learning_rate*db.mean()
        
    def predict(self, x):
        W1, W2, b = self.params
        return W1 * x**2 + W2 * x + b
```

```python
FOM = First_Order_Mom()
history = FOM.fit(x, y, 1000, 0.001, 0.5, batch=50)

x_test = np.linspace(0, 6, 50)
y_pred = custom_model.predict(x_test)

plt.figure(figsize=(15, 10))
plt.scatter(x, y, label="Data Point", marker=".")
plt.scatter(x_test, y_pred, label="Prediction")
plt.plot(x_test, y_pred, color="k", label="Fitted curve")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot([i for i in range(1, 1001)], history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

### Conclusiones: Se observa una convergencia más rápida en las primeras 100 epochs al emplearse el modelo de First order Momentum, aunque posteriormente, los errores decaen muy rápidamente y tienden asintóticamente el mismo valor, muy cercano a cero.
